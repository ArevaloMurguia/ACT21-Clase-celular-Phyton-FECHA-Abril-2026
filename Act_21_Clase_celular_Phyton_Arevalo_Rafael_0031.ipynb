```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

# Instanciación
cel1 = Celular("Samsung", "Galaxy S23", 800)
mi_cel = Celular("Apple", "iPhone 15", 1000)
# Llamada a las funciones para cada objeto cel1 y mi_cel
print("Objeto cel1:")
cel1.mostrar_info()
cel1.encender()

print("\nObjeto mi_cel:")
mi_cel.mostrar_info()
mi_cel.encender()


print("Programa terminado")
print("creador de software por Arevalo Rafael NL = 3")
```
